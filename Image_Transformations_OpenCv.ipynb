```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = '/content/Mountain img.jpg'
image = cv2.imread(image_path)
```

```python
alpha = 1.5
beta = 50
linear_transformed = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

plt.figure(figsize=(20, 10))
plt.subplot(2, 3, 2)
plt.imshow(linear_transformed, cmap='gray')
plt.title('Linear Transformation')
plt.axis('off')
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Bitwise Slicing
for bit_plane_index in range(1, 9):
    bit_plane = (gray_image >> bit_plane_index) & 1
    plt.figure(figsize=(10, 5))
    plt.subplot(2, 5, bit_plane_index)
    plt.imshow(bit_plane, cmap='gray')
    plt.title(f'Bit-Plane {bit_plane_index}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Thresholding
_, thresholded = cv2.threshold(image, 150, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(20, 10))
plt.subplot(2, 3, 3)
plt.imshow(thresholded, cmap='gray')
plt.title('Thresholding')
plt.axis('off')
```

```python
# Gray-Level Slicing
lower = 150
upper = 255
gray_level_sliced = cv2.inRange(gray_image, lower, upper)

plt.figure(figsize=(20, 10))
plt.subplot(2, 3, 5)
plt.imshow(gray_level_sliced, cmap='gray')
plt.title('Gray-Level Slicing')
plt.axis('off')
```
