Image Load and Read

```python
import cv2
import matplotlib
```

```python
Image1 = cv2.imread('/content/397483-1685429016.jpg')
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(Image1)
```

Image Bluring

```python
median_image = cv2.medianBlur(Image1, 5)
cv2_imshow(median_image)
```

Image Greyscaling

```python
gray_image = cv2.cvtColor(Image1, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

Threshold Defining

```python
ret, thresh_img = cv2.threshold(gray_image, 200, 255, cv2.THRESH_BINARY)
cv2_imshow(thresh_img)
```

```python
edged_img = cv2.Canny(thresh_img, 30, 200)
cv2_imshow(edged_img)
```

Counting Contours

```python
contours, _ = cv2.findContours(thresh_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

contour_count = 0

for contour in contours:
    contour_count += 1

print("Number of contours:", contour_count)
```

Hit and Trials

```python
ret, thresh_img2 = cv2.threshold(gray_image, 150, 255, cv2.THRESH_BINARY)
cv2_imshow(thresh_img2)
```
